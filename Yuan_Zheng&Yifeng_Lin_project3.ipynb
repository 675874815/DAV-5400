Analyzing Chess Tournament Results and get each Players'Name, Player’s State, Total Number of Points, Player’s Pre-Rating, and Average Pre Tournament Chess Rating of Opponents, then output them into the CSV.file

```python
#import the library of pandas, numpy and sys
import pandas as pd
import numpy as np
import sys
```

```python
#import the file of txt

result=[]
with open(r'C:\Users\14914\Desktop\tournamentinfo.txt','r') as f:
    for line in f:
        result.append(list(line.strip('\n').split(',')))
result[:4]
```

We find that the first 4 lines are useless to extract data, so we decide to drop them.

```python
#drop the first four lines of txt
result_new=result[4:]
result_new
```

We know some lines contain players' names, total points, and the number of opponents, so we are going to separate them into two parts.

```python
#calculate each parts' length
num1=len(result_new)/3
num2=int(num1)
num2
```

```python
#make up two lists for each parts' number
list1=[3*i for i in range(0,num2,1)]
list2=[3*i+1 for i in range(0,num2,1)]
```

```python
#separate the information about the player into two parts
player1=[]
player2=[]

for i in list1:
    player1.append(result_new[i])
    
for i in list2:
    player2.append(result_new[i])
    
player1[:10]

    
```

```python
player2[:10]
```

We found that there are a lot of '|' and other symbols in two lists, so we are going to figure out the index of each symbol, which will help us to find the data we want. 

```python
#define three lists, and put the index of symbols into them
index=[idx for idx,i in enumerate(player1[0][0]) if i=='|']
index2=[idx for idx,i in enumerate(player2[0][0]) if i==':']
index3=[idx for idx,i in enumerate(player2[0][0]) if i=='-']
```

```python
#extract the players' name 
player_name=[]
for i in range(len(player1)):
    player_name.append(str.strip(player1[i][0][index[0]+1:index[1]]))
player_name
```

```python
#extract the players' total points
player_total_points=[]
for i in range(len(player1)):
    player_total_points.append(str.strip(player1[i][0][index[1]+1:index[2]]))
player_total_points
```

```python
#extract players' states
player_state=[]
for i in range(len(player1)):
    player_state.append(str.strip(player2[i][0][:index[0]]))
player_state
```

```python
#extract players' previous rating
player_prerating_score=[]
for i in range(len(player1)):
    player_prerating_score.append(str.strip(player2[i][0][index2[0]+1:index3[0]]))
player_prerating_score
```

In order to calculate the  Average Pre Tournament Chess Rating of Opponents, firstly we need to find out each number of players' opponents, and then we need to add the pre-ratings of each opponent and divide the number of opponents, the result is what we want.

```python
#list out each index of "|"
index
```

find out each number of players' opponents, we need to use the index of each "|" to locate the number we what

```python
#Extract the number of each player's opponent
#The 2 and 9 in for loop represents the index of the thrid "|" and the last '|'
Player_avg_prescore=[]
for i in range(len(player1)):
    Player_avg_prescore.append([])
    for j in range(2,9):
            Player_avg_prescore[i].append(str.strip(player1[i][0][index[j]+2:index[j+1]]))
Player_avg_prescore   
```

There is a lot of missing value in this list and we need to first make each value into the type of integer, and then turn the NaN into zero, which would be better for analyzing.

```python
#clean the missing value to make it easier to analyze
for i in range(len(player1)):
    Player_avg_prescore[i]=pd.to_numeric(Player_avg_prescore[i])   

Player_avg_prescore
player3=np.array(Player_avg_prescore)
player4=np.nan_to_num(player3)
player4
```

Because there are some values in the list player_prerating_score, so we need to split it with 'P' and choose the first part.

```python
#use two 'for' loop to extract the average pre-rating of each opponents.
opponent_avg=[]
for i in range(len(player1)):
    sum=0
    num=0
    for j in range(7):
        if player4[i][j]!=0:
            sum=sum+int(player_prerating_score[int(player4[i][j]-1)].split('P')[0])
            num=num+1
            pre_avg=round(sum/num)
    opponent_avg.append(pre_avg)
    
opponent_avg[:10]

```

```python
#Combine these lists together and set up the name of each column
#Output the CSV format of this data frame
df = pd.DataFrame({'name':player_name, 'total_points':player_total_points,'state':player_state,'prerating_score':player_prerating_score,'opponent_avg':opponent_avg})  
df
df.to_csv(r'C:\Users\14914\Desktop\P3.csv')
```
