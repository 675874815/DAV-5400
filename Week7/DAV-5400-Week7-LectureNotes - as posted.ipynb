### What about scatter plots?

- Scatter plots should be used when your are trying to visualize the relationships between two or more __numeric__ variables.


### When can I use a scatterplot for a categorical variable?

- The short answer is "never"

- For categorical variables, we can use __strip plots__ or __swarm plots__ (both are not truly "scatter plots") to visualize the relationships __between__ a categorical variable's values and numeric values:

https://seaborn.pydata.org/tutorial/categorical.html


### What about line graphs?

- Line graphs are appropriate for __time series__ numeric data, i.e., we want a plot of how the value of one or more variables changes relative to a chronological reference/indicator.


### How about pie charts?

- Short answer: __never use pie charts for analytics work__. Pie charts are prone to misinterpretation due to the need for the human eye to try to discern differences in partial areas of a circle.

https://www.businessinsider.com/pie-charts-are-the-worst-2013-6

## Matplotlib

- __Matplotlib__ is a desktop plotting package used mostly for 2-d plotting.

- Designed to enable __MATLAB__-style plotting constructs within Python

- Enables very detailed plotting control options

- Supports most vector and raster graphics formats (e.g., PDF, SVG, JPG, PNG, etc.)

- The Pandas plotting functions we've been using recently were built using Matplotlib.

- Great tutorials are available here: https://matplotlib.org/tutorials/index.html


### Getting Started with Matplotlib

%matplotlib notebook

```python
# load the pyplot function from the matplotlib library
import matplotlib.pyplot as plt

# load numpy so that we can generate some sample data
import numpy as np
```

```python
# make a simple plot: randomly sample 50 items from a normal
# distribution and plot the cumulative sum
# 'k--' is a Matplotlib 'style' => plot a dashed black line
# the semicolon suppresses extraneous output that Matplotlib generates
plt.plot(np.random.randn(50).cumsum(), 'k--');
```

As we can see in the above plot, __Matplotlib__ automatically generates "tick" marks for both the X and Y axes for numeric data: the X axis represents the sample item counter while the Y axes represents the cumulative sum values.

```python
# plot a histogram
plt.hist(np.random.randn(50));
```

```python
# plot a box plot
plt.boxplot(np.random.randn(50));
```

### Matplotlib "Figures": Customize Your Graphics

- Every Matplotlib graphic resides within a Matplotlib "figure"

- We use "figures" instead of the basic plots shown above when we want to customize our plots (e.g., adding custom legends, titles, tick marks, color, etc.)

- Matplotlib Figures also provide a mechanism for generating subplots within a single plot panel.

#### Example for a Single Plot

```python
# define a Matplotlib figure
plt.figure(1)

# define the plot you want to generate
plt.hist(np.random.randn(100))

# define a title for the plot
plt.title('Histogram of 100 Samples from Normal Distribution')

# define the number of bins for the histogram + which plot axis they will
# be applied to
plt.locator_params(nbins=10, axis = 'x')

plt.show()
```

#### Subplots: Side-by-Side Plots

- Subplots are enabled by specifying an array-like dimensionality

- First define a new figure (e.g., "plt.figure(1)")

- Then specify how you want your subplots laid out within the figure using the "plt.subplot()" function

An example:

```python
# side-by-side boxplot and histogram of one set of data

data = np.random.randn(100)

# define a new figure for plotting
plt.figure(1)

# define the plot that will appear on the lefthand side of the subplot:
# the '121' value passed to the function says: "I am defining a 1x2 arrangement
# of subplots (as indicated by the first 2 digits '12'), and this is the 
# specification for the plot that will appear on the lefthand side (as indicated
# by the last digit '1')"
plt.subplot(121)

# define the type of plot you want to appear
plt.boxplot(data)

# give the plot a title
plt.title('Boxplot of Data')

# Here we'll blank out the x-axis label since it isn't really meaningful
# for a box plot
plt.xlabel(' ')

# --------------------------------------------------------------------------
# now define the subplot that will appear on the righthand side of our 1x2
# subplot ('122' = 1x2, position 2)
plt.subplot(122)

# define the type of plot you want to appear
plt.hist(data)

# give the plot a title
plt.title('Histogram of Data')

# specify the number of bins to use in the histogram
plt.locator_params(nbins=10, axis = 'x')

plt.show()
```

### Color, Markers, and Line Styles

### Ticks, Labels & Legends

```python
# define a new figure
fig = plt.figure()
# specify that only a single subplot will be placed within the figure
ax = fig.add_subplot(1, 1, 1)

# plot cumulative sum of 1000 random samples taken from normal distrib.
ax.plot(np.random.randn(1000).cumsum())

# specify where the tick marks for the x axis will be placed
ticks = ax.set_xticks([0, 250, 500, 750, 1000])

# specify labels for the tick marks on the x axis
# 'rotation' lets you rotate the tick labels (default is horizontal)
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                            rotation=30, fontsize='small')

# define a title for the plot: gets displayed at the top
ax.set_title('Example from Page 267-8 in PfDA Text')

# define a label for the plot: gets displayed at the base of the plot
ax.set_xlabel('Stages');
```

#### To customize Y axis ticks + labels, just substitute 'y' for 'x' in the above function calls

Example of adding a legend (pp. 269-270 in PfDA text):

```python
fig = plt.figure(); 
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best');
```

## Seaborn

- Seaborn is a Python library for generating statistical graphics.

- Seaborn is basically a "wrapper" around Matplotlib: it tries to make the generation of certain types of plots more "user friendly"

### Getting Started with Seaborn

```python
import seaborn as sns
```

```python
# make a simple plot: randomly sample 50 items from a normal
# distribution and plot the cumulative sum
# the semicolon suppresses extraneous output that Matplotlib generates

sns.lineplot(data = np.random.randn(50).cumsum())
```

```python
# plot a histogram
sns.distplot(np.random.randn(1000), bins = 10)
```

```python
# plot a box plot
sns.boxplot(np.random.randn(50), orient = 'v');
```

### Side-by-Side Plots Using Seaborn

- Use the Matplotlib plt.figure() constructs to define your plotting space

- Then simply use Seaborn plotting in place of Matplotlib plotting

```python
# side-by-side boxplot and histogram of one set of data
# the procedure is same as above for Matplotlib but uses
# Seaborn plotting functions instead of Matplotlib functions.

data = np.random.randn(1000)

# define a new figure for plotting
plt.figure(1)

# define the plot that will appear on the lefthand side of the subplot:
# the '121' value passed to the function says: "I am defining a 1x2 arrangement
# of subplots (as indicated by the first 2 digits '12'), and this is the 
# specification for the plot that will appear on the lefthand side (as indicated
# by the last digit '1')"
plt.subplot(121)

# define the type of plot you want to appear
# plt.boxplot(data) # <= Matplotlib boxplot function
sns.boxplot(data, orient = 'v');

# give the plot a title
plt.title('Boxplot of Data')

# Here we'll blank out the x-axis label since it isn't really meaningful
# for a box plot
plt.xlabel(' ')

# --------------------------------------------------------------------------
# now define the subplot that will appear on the righthand side of our 1x2
# subplot ('122' = 1x2, position 2)
plt.subplot(122)

# define the type of plot you want to appear: set the number of bins
# for the x axis to 10
# Note that seaborn uses'distplot' for histograms, while matplotlib
# use 'hist':
# plt.hist(data) <= Matplotlib histogram function
sns.distplot(data, bins = 10)

# give the plot a title
plt.title('Histogram of Data')

plt.show();
```

### Scatter/Point Plots in Seaborn

```python
# use sample data from PfDA text
macro = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/macrodata.csv')

# take a subset of the macro data frame
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]

# do some calulations
trans_data = np.log(data).diff().dropna()

# display the last 5 rows
trans_data[-5:]
```

```python
# scatterplot of cpi vs. unemp
sns.scatterplot(x = 'cpi', y = 'unemp', data = trans_data)
```

### Regression Plots in Seaborn

```python
# generate a regression plot using Seaborn

# first define a Matplotlib figure
plt.figure()

# use Seaborn's regplot() function to plot the regression line for the 
# variables m1 and unemp from the trans_data dataframe
sns.regplot('m1', 'unemp', data=trans_data)

# give the plot a title using Matplotlib
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

### Pair Plots in Seaborn

```python
# automatically generate a pairs plot using Seaborn: each variable
# within the trans_data data frame is plotted against every other 
# variable within the data set.
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

### Facet Grids in Seaborn

- Facet grids are useful if you are plotting categorical data. In the example below we are barplotting tip percentages based on whether the customer was a smoker or a non-smoker. Seaborn automatically separates the data based on the value indicated in the "smoker" column

- Facet grids are a tool often used in Exploratory Data Analysis (EDA) work

```python
# NOTE: the 'factorplot()' function has been renamed to "catplot()"
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# Add another categorical: Now we are faceting based on both mealtime AND
# smoker/nonsmoker
sns.catplot(x='day', y='tip_pct', row='time',
               col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# a faceting example using boxplots instead of barplots
sns.catplot(x='tip_pct', y='day', kind='box',
               data=tips[tips.tip_pct < 0.5])
```

```python
import pandas as pd
import numpy as np

# load the data set
# auto_df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", delim_whitespace = True, header = None)
# Posted copy of data set to Github since UCI no longer provides direct access
auto_df = pd.read_csv("https://raw.githubusercontent.com/jtopor/DAV-5400/master/Week7/auto-mpg.data", delim_whitespace = True, header = None)
```

```python
auto_df.head(10)
```

```python
# add meaningful column names
auto_df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',
                   'weight', 'acceleration', 'model', 'origin', 'car_name']

# sanity check
auto_df.head(5)
```

```python
# how many '?' values are there in the horsepower column?
auto_df.horsepower.value_counts()['?']
```

```python
# how many 'NaN' values are in the horsepower column?
auto_df.horsepower.isna().sum()
```

```python
# replace '?' in horsepower column with 'NaN'
auto_df.horsepower.replace('?', np.nan, inplace = True)
```

```python
# how many 'NaN' values are there in the horsepower column?
auto_df.horsepower.isna().sum()
```

```python
# convert the column to numeric
auto_df["horsepower"] = pd.to_numeric(auto_df["horsepower"])
```

```python
# replace origin values using a dict
auto_df.origin.replace({1: 'USA', 
                             2: 'Asia',
                             3: 'Europe'}, inplace = True)
```

```python
auto_df.head(10)
```

```python
# create a graphic showing the distribution for cylinders
sns.distplot(auto_df["cylinders"]);
```

```python
# scatterplot of horsepower vs. weight
sns.scatterplot(x = 'horsepower', y = 'weight', data = auto_df);
```

```python
# fit a linear model to the scatter plot
sns.lmplot(x = 'horsepower', y = 'weight', data = auto_df);
```

```python
# strip plot of mpg vs. model year: note the upward trend in mpg
sns.scatterplot(x = 'model', y = 'mpg', data = auto_df);
```

```python
# facet grid for model year vs. mpg by point of origin
# NOTE: the 'factorplot()' function has been renamed to "catplot()"

# note how the general trend in mpg is upward for each point of origin
sns.catplot(x='model', y='mpg', col='origin',
               kind='bar', data=auto_df);
```

```python
# facet grid for model year vs. weight by point of origin
# NOTE: the 'factorplot()' function has been renamed to "catplot()"

# note how the general trend in mpg is downward for cars manufactured in USA
# but does not appear to be trending downward for vehicles produced in Europe
# or Japan
sns.catplot(x='model', y='weight', col='origin',
               kind='bar', data=auto_df);
```
